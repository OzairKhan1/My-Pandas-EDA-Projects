```python
# Now I want to grab the duplicates Name of the Channel if any
df['Channel name'].duplicated().any()     # Yes the duplicates are available
```

```python
# Now I need to find how many of them are duplicates
df['Channel name'].duplicated().sum()           # 7 duplicates are present in the data
```

```python
# Now I want to view those duplicates
dupNames = df[df['Channel name'].duplicated(keep = False)]     # The keep = False is used to show all the instances of duplicates
dupNames
```

```python
# Now i want to sort them on the basis of channel names
dupNames = dupNames.copy()
dupNames.sort_values(by = ['Channel name', 'Subscribers'],inplace = True)
dupNames
```

```python
dupNames['Subscribers'] = dupNames['Subscribers'].replace('--','0',regex = True)
dupNames['Subscribers'] = dupNames['Subscribers'].astype(int)
```

```python
# Now I want to see the graph
plt.figure(figsize=(10,5))
ax = sns.barplot(x = 'Rank', y = 'Subscribers', data = dupNames, hue = 'Channel name')
plt.xticks(rotation = 25)
for labels in ax.containers:
    ax.bar_label(labels)
plt.yscale('log')
plt.legend(loc = (1,0.3))
plt.title('Un Sorted Duplicated Channel Name Subscribers by Rank',size = 10)
plt.show()

# In the Fig below the data is not sorted on the subscriber basis, but in the fig 2 the data is sorted
```

```python
SortedDupNames = dupNames.sort_values(by='Subscribers', ascending=False)

plt.figure(figsize=(10, 6))
ax = sns.barplot(x='Rank', y='Subscribers', data=SortedDupNames, hue='Channel name')
plt.xticks(rotation=25)
plt.xlabel('Rank')
plt.ylabel('Subscribers')
plt.title('Sorted Duplicated Channel Name Subscribers by Rank',size = 10)
for labels in ax.containers:
    ax.bar_label(labels)
plt.legend(loc = 'center right')
plt.show()
```

```python
# 6: replace the -- by NaN
df = df.replace('--',np.nan,regex = True)
```

```python
# 7; Check the null values and delete them if more 
print(df.isna().any())          # To check whether available or not
```

```python
# df.isna().sum()
# or to see the percentage 
perc = df.isnull().sum()/len(df) * 100
perc
```

```python
# To Visualize them
sns.heatmap(df.isnull())
```

```python
# Lets drop the na values
df.dropna(axis = 0,inplace = True)
```

```python
print(df.isna().sum())
plt.figure(figsize=(3,2))
sns.heatmap(df.isna())      # Now there is not null values
```

```python
# 8: Now WE need to clean the Rank Cols
df['Rank'] = df['Rank'].str[0:-2]
```

```python
df['Rank'] = df['Rank'].str.replace(',','')
```

```python
df['Rank']
```

```python
df = df.astype({'Video Uploads':'int', 'Subscribers':int})   # Here the data types has changed from object to int
```

```python
# 10: Grade Column Cleaning
# We need to convet or map it with integers so that we can use it for Ml Probjects later on
df['Grade'].unique()
# these are the unique data in the dataset
```

```python
df['Grade'] = df['Grade'].map({'A++ ':5, 'A+ ':4, 'A ':3, 'A- ':2, 'B+ ':1}) 
```

```python
df['Grade'].unique() # After Mapping. 
```

```python
df['Avg_Views'] = df['Video views']/df['Video Uploads']
```

```python
df.head()
```

```python
top_10 = df.head(10)
```

```python
top_10
```

```python
sortedTop10 = top_10.sort_values(by = 'Subscribers', ascending = False)
ax = sns.barplot(y = 'Channel name', x = 'Subscribers',data = sortedTop10, hue = 'Rank', palette = 'Set1' )
for labels in ax.containers:
    plt.bar_label(labels)
plt.title("Top 10 Channel Vs Subscribers group by Rank ")
plt.show()
```

```python
# Find out 5 Channel with maximum number of vides upload

df.sort_values(by = 'Video Uploads',ascending = False).head()['Channel name']
```

```python
# 13: Find the correlation Matrix: We can't perform the Correlatin on object or string. so we have to chools only int or float
numeric_df = df.select_dtypes(include=[float, int])
numeric_df.corr()
```

```python
# 14: Which grade has maximum number of uploads
print(df.groupby('Grade')['Video Uploads'].mean())
ax = sns.barplot(x = 'Grade', y = 'Video Uploads', data = df)
for labels in ax.containers:
    ax.bar_label(labels)
plt.show()
```

```python
# 15: # which grade has higher avg views

print(df.groupby('Grade')['Avg_Views'].mean())
ax = sns.barplot(x = 'Grade', y = 'Avg_Views', data = df, color = 'lightblue')
for labels in ax.containers:
    ax.bar_label(labels)
plt.title("Grade Vs Avg Views")
plt.show()
```

```python
print(df.groupby('Grade')['Subscribers'].mean())
ax = sns.barplot(x = 'Grade', y = 'Subscribers', data = df, color = 'lightgreen')
for labels in ax.containers:
    ax.bar_label(labels)
plt.show()
```

```python
dfc = df.groupby('Grade')['Video views'].mean()
print(dfc)
explode = [0,0,0,0.2,0]
plt.pie(dfc, labels= [1,2,3,4,5], autopct='% 0.2f %%',shadow=True, startangle = -40, frame = True, explode =explode,pctdistance=0.6)
plt.title("Grade Vs Video Views")
plt.show()
```
